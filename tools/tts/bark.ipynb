```python
from transformers import BarkModel

model = BarkModel.from_pretrained("suno/bark-small")
```

```python
import torch

device = "cuda:0" if torch.cuda.is_available() else "cpu"
model = model.to(device)
```

```python
from transformers import AutoProcessor

processor = AutoProcessor.from_pretrained("suno/bark")
```

```python
sampling_rate = model.generation_config.sample_rate
```

```python
#voice_preset = "v2/en_speaker_9"
voice_preset = "v2/ko_speaker_0"

# prepare the inputs
text_prompt = "안녕하세요 만나서 반갑습니다"
inputs = processor(text_prompt, voice_preset=voice_preset)

# generate speech
speech_output = model.generate(**inputs.to(device), temperature = 0.1, semantic_temperature = 0.1)

```

```python
from IPython.display import Audio

# let's hear it
Audio(speech_output[0].cpu().numpy(), rate=sampling_rate)
```

```python
import scipy

scipy.io.wavfile.write("bark_out.wav", rate=sampling_rate, data=speech_output[0].cpu().numpy())
```
